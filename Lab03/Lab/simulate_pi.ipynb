# Simulating $\pi$

The area of a circle $C$ is given as $|C| = \pi r^2$ with $r$ as the radius.

A square $S$ has area $|S| = w^2$ where $w$ is the length of a side of the square.

If we put a circle with radius $r$ inside of a square with edge $w=2r$ then the ratio of the areas is:

$$ratio = \frac{|C|}{|S|} = \frac{\pi r^2}{w^2} = \frac{\pi r^2}{(2r)^2}= \frac{\pi r^2}{4r^2} = \frac{\pi}{4} $$

Thus if we multiply the ratio of the areas by 4 we have a formula for $pi$:
$$ \pi = 4 ratio$$ 
or
$$ \pi = 4\frac{|C|}{|S|}$$



We can use this relationship to simulate the ratio of the areas to generate an estimate of $\pi$.

Basically we take uniform random draws of an $x$ and $y$ coordinate for a point and look at the share of many such points that fall within the circle of all those that are in the square.

We want to get within $p=0.00001$ of the value of $\pi$ that is reported in Python and we are curious as to how many points we have to draw to achieve this level of accuracy.

```python
import random
import math
import matplotlib.pyplot as plt
```

```python
pi = math.pi # our benchmark that we want to estimate using simulation
```

## random.random()
Return the next random floating point number in the range [0.0, 1.0).

```python
n = 0                                  # number of points falling in the unit circle
d = 0                                  # number of points falling in the unit square
simulating = True                      # use as a sentinel
while simulating:
    x = random.random()                # represents a random point on the x-axis bounded by the unit square
    y = random.random()                # represents a random point on the y-axis bounded by the unit square
    if x**2 + y**2 <= 1.0:             # by Pythagoras theorem, any point (x, y) lies inside the unit circle if the sum of the squares of the coordinates is less than or equal to 1
        n += 1                         # if the point lies inside the circle, n is incremented
    d += 1                             # total number of points d is incremented with every iteration, regardless of whether they fall inside the circle
    ratio1 = 4 * n * 1./d               # gives an estimate of π based on the fraction of points inside the circle out of the total points (d)
#    print(ratio1)
    if abs(ratio1-pi) / pi <= 0.00001:  # loop continues to generate random points and calculate the ratio until the estimated value of π is accurate to within 0.00001 of the actual value of π
        print("Draws needed: ", d)
        break
```

```python
n = 0                                  # number of points falling in the unit circle
d = 0                                  # number of points falling in the unit square
ratios = []                            # a list that stores the estimate of π at each iteration
xs = []                                # a list that stores the x coordinates of the randomly generated points
ys = []                                # a list that stores the y coordinates of the randomly generated points
simulating = True                      # use as a sentinel
while simulating:
    x = random.random()                # represents a random point on the x-axis bounded by the unit square
    y = random.random()                # represents a random point on the y-axis bounded by the unit square
    xs.append(x)                       # append newly generated x coordinate to the list xs, storing the x coordinate of every point generated so far
    ys.append(y)                       # append newly generated y coordinate to the list ys, storing the y coordinate of every point generated so far
    if x**2 + y**2 <= 1.0:             # by Pythagoras theorem, any point (x, y) lies inside the unit circle if the sum of the squares of the coordinates is less than or equal to 1
        n += 1                         # if the point lies inside the circle, n is incremented
    d += 1                             # total number of points d is incremented with every iteration, regardless of whether they fall inside the circle
    ratio2 = 4 * n * 1./d               # gives an estimate of π based on the fraction of points inside the circle out of the total points (d)
    
    print(ratio2)                       # estimate is printed in real-time during each iteration
    ratios.append(ratio2)               # ratio for each iteration is appended to the ratios list 
    if abs(ratio2-pi) / pi <= 0.00001:  # loop continues to generate random points and calculate the ratio until the estimated value of π is accurate to within 0.00001 of the actual value of π
        print("Draws needed: ", d)
        break
```

Extend the Python notebook to prompt the user for two input values:

- The convergence criterion (i.e., percentage difference for stopping the loop)
- A sentinel value that limits the number of draws to avoid a long running program. Once the number of draws reaches this value you should have the program stop and print a message that draw limit was exceeded and report the percentage difference between the current estimate of Pi and the true value.

```python
# ask the user for the convergence criterion (as a percentage)
tolerance_percentage = float(input("Enter the convergence criterion as a percentage (e.g., 0.001 for 0.001%): "))

# convert the percentage into a decimal tolerance value for the stopping condition
tolerance = tolerance_percentage / 100

# ask the user for a sentinel value (limit for number of draws)
max_draws = int(input("Enter the maximum number of draws: "))

n = 0                                           # number of points falling in the unit circle
d = 0                                           # number of points falling in the unit square
ratios_Lab03_2 = []                             # a list that stores the estimate of π at each iteration
xs = []                                         # a list that stores the x coordinates of the randomly generated points
ys = []                                         # a list that stores the y coordinates of the randomly generated points
simulating = True                               # use as a sentinel
while simulating:
    x = random.random()                         # represents a random point on the x-axis bounded by the unit square
    y = random.random()                         # represents a random point on the y-axis bounded by the unit square
    xs.append(x)                                # append newly generated x coordinate to the list xs, storing the x coordinate of every point generated so far
    ys.append(y)                                # append newly generated y coordinate to the list ys, storing the y coordinate of every point generated so far
    if x**2 + y**2 <= 1.0:                      # by Pythagoras theorem, any point (x, y) lies inside the unit circle if the sum of the squares of the coordinates is less than or equal to 1
        n += 1                                  # if the point lies inside the circle, n is incremented
    d += 1                                      # total number of points d is incremented with every iteration, regardless of whether they fall inside the circle
    ratio_Lab03_2 = 4 * n * 1./d                # gives an estimate of π based on the fraction of points inside the circle out of the total points (d)
    
    print(ratio_Lab03_2)                        # estimate is printed in real-time during each iteration
    ratios_Lab03_2.append(ratio_Lab03_2)        # ratio for each iteration is appended to the ratios list

    # Stopping condition using user-defined tolerance
    if abs(ratio_Lab03_2-pi) / pi <= tolerance: # loop continues to generate random points and calculate the ratio until the estimated value of π is accurate to within the convergence criterion of the actual value of π
        print("Draws needed: ", d)
        break

    # Sentinel to stop the programme after reaching the maximum number of draws
    if d >= max_draws:
        percent_difference = abs(ratio_Lab03_2-pi) / pi * 100 # calculate percentage difference
        print(f"Draw limit of {max_draws} exceeded.")
        print(f"Percentage difference between estimated π and true π: {percent_difference}%")
        break
```
